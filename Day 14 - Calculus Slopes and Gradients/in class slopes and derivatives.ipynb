```python
x = np.linspace(-3,10,100)
 
y = x**2
 
result = np.array(y)
 
 
plt.plot(x,result)
plt.axhline(0, color = 'black', linewidth = 0.4)
plt.axvline(0, color = 'black', linewidth = 0.4)
```

```python
x = np.linspace(-5,5,50)
 
e = 2*x**2 + 3
 
 
plt.plot(x,e, color = 'purple')
plt.axhline(0, color = 'black', linewidth = 0.4)
plt.axvline(0, color = 'black', linewidth = 0.4)
plt.show()
```

```python
x = np.linspace(-10,5,100)
 
e = 2*(x+3)**2 + 3
 
 
plt.plot(x,e, color = 'purple')
plt.axhline(0, color = 'black', linewidth = 0.4)
plt.axvline(0, color = 'black', linewidth = 0.4)
plt.show()
```
