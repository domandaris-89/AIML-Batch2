**Exercise 1 - Add Tip Percentage**

    1. Create a new column:
        tip_pct = tip / total_bill *100
    2. Compute correlation of tip_pct wtih total_bill and with size.
    3. Look at the heatmap again. What changed?


```python
dftips.head()
```

```python
tip_pct =  (dftips['tip'] / dftips['total_bill']) * 100
dftips['tip_pct'] = tip_pct

dftips.head()
```

```python
columns = ['total_bill','tip','size','tip_pct']
tips_corr = dftips[columns].corr()
tips_corr.round(2)
```

```python
sns.heatmap(tips_corr, cmap = 'coolwarm', annot=True, fmt='.2f', vmin= -1, vmax= 1)
```
