```python
# data
# bar graph

# create a list of students
#grades

students=['idrees','ibrahim','mel','hazirah','amin','nazri']
grades = [38,95,95,55,97,98]
```

```python
import matplotlib.pyplot as plt

# create a bar plot
plt.bar(students,grades, color = 'pink') # specify properties of our bar chart

plt.title('Chart of student grades')

plt.xlabel('student names')

plt.ylabel('marks')

plt.show() #you might need this
```

```python
# create a scatter plot
plt.scatter(students,grades, color = 'purple') # specify properties of our bar chart

# prettying up the visual
plt.title('Chart of student grades')
plt.xlabel('student names')
plt.ylabel('marks')
plt.show() #you might need this
```

```python
# create a line chart/plot
plt.plot(students,grades, color = 'purple') # specify properties of our bar chart

# prettying up the visual
plt.title('Chart of student grades')
plt.xlabel('student names')
plt.ylabel('marks')
plt.show() #you might need this
```

```python
# all of them in the same place

# create a figure of 1 row, and 3 columns
plt.figure(figsize=(15,5))

#sub plot

plt.subplot(1,3,1) # rows, columns position
plt.bar(students,grades, color = 'pink') # specify properties of our bar chart

plt.subplot(1,3,2) # rows, column second
plt.plot(students,grades, color = 'purple') # specify properties of our bar chart

plt.subplot(1,3,3) # rows, column third
plt.scatter(students,grades, color = 'indigo') # specify properties of our bar chart
```

```python
cities=['KL','SG','NY','HK','BL']
temp=[29,31,19,27,16]
```

```python
plt.bar(cities,temp, color = 'green') # specify properties of our bar chart
plt.title('Chart of cities temperatures')
plt.xlabel('cities')
plt.ylabel('temperatures in C')
plt.show()
```
